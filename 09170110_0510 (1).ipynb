```python
from PIL import Image,ImageDraw,ImageFont
from google.colab.patches import cv2, cv2_imshow 
pic=Image.new("RGB",(700,500),"#CCFF80")
draw_pic=ImageDraw.Draw(pic)
draw_pic.line([(0,0),(699,0),(699,499),(0,499),(0,0),(0,250),(700,250)],fill="black")

txt="學校網頁:"
txt1="我的名片:"
fontinfo=ImageFont.truetype("/content/msjh.ttc",30)
draw_pic.text((50,50),txt,fill="olive",font=fontinfo)
draw_pic.text((50,300),txt1,fill="olive",font=fontinfo)

pp=Image.open("44.jpg")
pic.paste(pp,(500,50))
pp1=Image.open("55.jpg")
pp1=pp1.resize((100,100))
pic.paste(pp1,(500,300))
pic.save("11.jpg")
final=cv2.imread("11.jpg")
cv2_imshow(final)
```
