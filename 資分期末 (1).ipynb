```python
import requests,json
import matplotlib.pyplot as plt

url="https://opendata.cwb.gov.tw/api/v1/rest/datastore/F-D0047-091?Authorization=rdec-key-123-45678-011121314"
res=requests.get(url)
resjson=json.loads(res.text)

for i in range (0,22):
    print(resjson['records']['locations'][0]['location'][i]['locationName']+":")
    aa=resjson['records']['locations'][0]['location'][i]['weatherElement'][1]['time']
    for i in aa:
        for o in i['elementValue']:            
            print(i["startTime"],i["endTime"],o['value']+"°C")

print("----------------------------------------------")
taipei=[]
print(resjson['records']['locations'][0]['location'][9]['locationName']+":")
aa=resjson['records']['locations'][0]['location'][9]['weatherElement'][1]['time']
for i in aa:
    for o in i['elementValue']:            
        taipei.append(int(o['value']))
print(taipei)
taichung=[]
print(resjson['records']['locations'][0]['location'][20]['locationName']+":")
aa=resjson['records']['locations'][0]['location'][20]['weatherElement'][1]['time']
for i in aa:
    for o in i['elementValue']:            
        taichung.append(int(o['value']))
print(taichung)
tainan=[]
print(resjson['records']['locations'][0]['location'][6]['locationName']+":")
aa=resjson['records']['locations'][0]['location'][6]['weatherElement'][1]['time']
for i in aa:
    for o in i['elementValue']:            
        tainan.append(int(o['value']))
print(tainan)

print("----------------------------------------------")
l=[]
def d(x):
    aa=resjson['records']['locations'][0]['location'][x]['weatherElement'][1]['time']
    for i in aa:
        for o in i['elementValue']:            
            l.append(int(o['value']))
            result=format(sum(l)/len(l),".1f")
    return result


```
